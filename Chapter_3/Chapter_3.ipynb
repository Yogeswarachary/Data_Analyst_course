##### 3. Predictive modeling

3.1 Introduction to predictive modeling

> Predictive modeling is making use of past data and trying to predict the future the future data with help of the past data.

3.2 Predictive Modeling types
> a) Supervised Learning
> b) Semisupervised Learning

A) Supervised Learning
> In this we have to train the dataset. so that we can use the labeled dataset. Always remember we are using the training dataset.
eg alg:-- Support vector machine, Linear regression. LOgistic regression ect

B)Unsupervised Learning or Semisupervised learning
> In this we do not give any training data. Algorithm it self it can detect the hiddien patterns contained in the dataset.
eg alg:-- K-means clustering, KNN, Heirarchal clustering etc.

##### Understanding and Treating and The missing values

> Here is the example problem of Treating the missing values

```python
#importing libraries
import pandas as pd

#loding the dataset
file = pd.read_excel(r"D:\Excel\releted to datascience excel files\supermarket_sales.xlsx")
file.head()
```

```python
#printing the shape of the dataset
file.shape
```

```python
#print the details of the missing values using the Describe()
file.describe()
```

```python
#Identifying the missing values with hepl of isnull()
file.isnull()
```

```python
#doing the sum of all the null values in the dataset
file.isnull().sum()
```

```python
#deleting the all the null values in the dataset
file.dropna()
```

```python
#confiming the deletung the null values
file.dropna().isnull().sum()
```

Alright we are successfully deleted the null values

```python
#delete all the null values presented rows
file.dropna(how = "all")
```

```python
#checking the shape of the data and now the dataset does not have any null values
file.dropna(how = 'all').shape
```

```python
#deleting the column that is having the missing values
file.dropna(axis=1)
```

```python
#filling these missing values with zero (0)
file.fillna(0)
```

fill the MODE value of categorical data column in the dataset. if the column is having the continuous data, then fill this column with the MEAN or MEDIAN values of the that column

```python
#filling the missing values of the Outlet_size column with the MODE value
file["Outlet_Size"].fillna(file["Outlet_Size"].mode())
```

```python
#filling the missing values of the Item_weight column with the mean value
file["Item_Weight"].fillna(file["Item_Weight"].mean())
```

```python
#filling the misssng values of the Item_MRP column with the median values
file["Item_MRP"].fillna(file["Item_MRP"].mean())
```

###### Variable Tranceformation

> Here is the example program for the variable tranceformation

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#loading the dataset
df = pd.read_excel(r"D:\Excel\releted to datascience excel files\supermarket_sales.xlsx")
df.head()
```

```python
#plotting the item mrp in the histogram
df["Item_MRP"].plot.hist()
```

```python
#Use the Logarithmic method for variable tranceformation
np.log(df["Item_MRP"]).plot.hist()
```

```python
#Use the Square_root method for variable tranceformation
np.sqrt(df["Item_MRP"]).plot.hist()
```

```python
#covert the data from continuous to categorical data by using binning concept
bins=[0,15, 80]
group = ["Lowest", "highest"]
```

```python
#storing the above groups into another variable
df["Item_MRP"] = pd.cut(df["Item_MRP"],bins,labels = group)
```
